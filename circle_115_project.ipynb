````python
# A simple example
import numpy as np
import matplotlib.pyplot as plt

# np, plt are just an alias, it can be anything

x = np.linspace(0, 2, 100)

plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Simple Plot")

#plt.legend()

plt.show()
````

**PLOT TYPES**

> <u>Pairwise Data</u> - plot of pairwise (x, y), tabular & functional data (Plot, Scatter, Bar, Stem, Fill_between, Stackplot, Stairs)

> <u>Statistical Distribution</u> - Distribution of at least one variable (Hist, Boxplo, Errorbar, Violinplot, Eventplot, Hexbin, Pie, ecdf)

> <u>Gridded Data</u> - Arrays & images (imshow, pcolormesh, contour, barbs, quiver, streamplot)

> <u>Irregularly Gridded Data</u> - Plots of data Z<sub>x,</sub><sub>y</sub> on unstructured grids (tricontour, tripcolor, triplot)  

> <u>3D & Volumeric Data</u> - Plots of three dimensional xyz (scatter, plot_surface, plot_trisurf, voxels, plot_wirefram)

**Matplotlib** graphs your data on **Figures**, each of which can contain one or more **Axes** (an area where points can be specified in terms of x-y coordinates).

**pyplot.subplots** is the most commonly used interface. It create Figures with an Axes then **Axes.plot** is for drawing some data on the Axes.

**pyplot interface** serves two purposes,
1. Spins up the backend and keep track of GUI windows
2. It serves as a global state for Axes & Artists that allows a short-form API to plotting methods.

````python
# Example2 to explain how this works
fig, ax = plt.subplots() # This simply means to a figure with a single axes

# To display the information on the axes visually
ax.plot([1, 2, 3, 4], [1, 5, 3, 2], [2, 3, 5, 7])

# To give it a title
plt.title('Figure')

# To show it graphically
plt.show()
````

**A Figure** keeps track of all the child Axes.

**An Axes** is an Artist attached to a figure that containes a region for ploting data (Two or more Axis).

````python
import pandas as pd

data = r"/content/train.csv"
df = pd.read_csv(data)
df.head(10)
````

````python
# Taking a closer look at all the columns
df.columns
````

````python
# To ascertain the no of survival against their Pclass using plt plot

X = df['Survived'].values
Y = df['Pclass'].values

A = df['Age'].values
S = df['Sex'].values
````

From the plot above, there's no much correlation. This means that the survival rate was not accrding to class.

Then, let's see the correlation of all series.

````python
df.corr()
````

````python
# Distribution of survival by gender
survBySex = df.groupby('Sex')['Survived'].mean()
survBySex
````

````python
# We will use plot here
# It's syntax is: Axes.plot(*args, scalex=True, scaley=True, data=None, **kwargs)
# It plots x v y as lines or marks


plt.plot(survBySex, label='linear')

plt.show()
````

````python
# Using pie also on same output
# Pie plot has so many parameters
# Axes.pie(x, explode=None, labels=None, color=None etc)
# The fractional area of each wedge in array x is given by x/sum(x)

labels = ['Female', 'Male']
plt.pie(survBySex, labels=labels)

plt.show()
````

````python
# Distribution of Survival by passenger class

survByClass = df.groupby('Pclass')['Survived'].mean()
survByClass
````

````python
# Making the color of our pie plot none, Matplotlib will automatically assign colors to our output as shown
labels = [1, 2, 3]
plt.pie(survByClass, labels=labels)

plt.show()
````

````python
# How does the survival rate vary for different age groups (e.g., children, adults, seniors)?

age_groups = pd.cut(df['Age'], bins=[0, 18, 35, 60, float('inf')],
                    labels=['Child', 'Adult','Middle Aged' ,'Senior'])
ageGrp = df.groupby(age_groups)['Survived'].mean()
ageGrp
````

````python
plt.plot(ageGrp, label='linear')

plt.show()
````

The slop above showed us that survival rate of children was the highest followed by middled aged.
